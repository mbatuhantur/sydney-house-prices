```python
# Listeyi String türüne çevirmek için kullandığım hazır kod.
def listToString(s):

  #initialize on empty string
  str1 = " "

  #return string
  return (str1.join(s))

driver_path = r"C:\Users\Batuhan\Desktop\pr\Program Setup\chromedriver.exe"
browser= webdriver.Chrome(executable_path= driver_path)
browser.get("https://www.emlakjet.com/satilik-konut/bursa/")

ev_linkler = []

i = 1
while i <= 39:
    #try:
        ev_linkler.append(browser.find_elements_by_class_name("_2UE6v")[i-1].get_attribute("href"))
        i += 1
                
for i in ev_linkler[1::2]:
    browser.get(i)
    elements = browser.find_elements_by_css_selector("._2xeaw")
    fiyatlar = browser.find_elements_by_css_selector("._34630")
    
    detaylar = []
    fiyat = []
  
    for i in fiyatlar:
        print(i.text)
        fiyat.append(i.text)
    for i in elements:
        print(i.text)
        detaylar.append(i.text)
    det_str = listToString(detaylar)
    ayrı = det_str.split("\n") 
    df = pd.DataFrame(ayrı)
    df_yeni = df.iloc[7:53]
    df_yeni.iloc[45] = df_yeni.iloc[45].str[0:3]   
    df_yeni = df_yeni.reset_index()
    df_yeni.drop("index", axis = 1, inplace = True)
    df_liste = df_yeni.values.tolist()

    icerikler = []
    i = 1
    while i <= 45:
        print(df_liste[i])
        icerikler.append(df_liste[i])
        i = i+2
    net_fiyat = fiyat[1].strip()
    net_fiyat = net_fiyat.replace("TL","")
    icerikler.append([net_fiyat])
  
    df_icerikler = pd.DataFrame(icerikler).T 
    df_icerikler.to_csv(r"yeni.csv", encoding="utf-8", index=False,mode="a" )    
    i = i+1
    
        
```
